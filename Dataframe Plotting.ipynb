## A

- cylinders: Quantitative (Discrete)
- displacement: Quantitative (Continuous)
- horsepower: Quantitative (Continuous)

- year: Qualitative (Categorical)
- origin: Qualitative (Categorical) - 1 for USA, 2 for Europe, and 3 for Asia
- name: Qualitative (Categorical)

## B

```python
predictors_quantitative = (
    auto['mpg'],
    auto['cylinders'],
    auto['displacement'],
    auto['horsepower'],
    auto['weight'],
    auto['acceleration']
)

print(predictors_quantitative)
```

```python
min_values = {}
max_values = {}
ptp_values = {}

for column, values in zip(auto.columns, predictors_quantitative):
 
    numeric_values = pd.to_numeric(values, errors='coerce')
       
    valid_values = numeric_values.dropna()
    
    range_value = np.ptp(valid_values)  # Peak-to-peak (maximum - minimum)   
    min_value = np.min(valid_values)
    max_value = np.max(valid_values)
    
    ptp_values[column] = range_value
    min_values[column] = min_value
    max_values[column] = max_value

print('Range values for each predictor:\n', ptp_values)    
print('\nMinimum values for each predictor:\n', min_values)
print('\nMaximum values for each predictor:\n', max_values)
```

## C

```python
mean_values = {}
standard_deviation_values = {}

for column, values in zip(auto.columns, predictors_quantitative):
 
    numeric_values = pd.to_numeric(values, errors='coerce')
       
    valid_values = numeric_values.dropna()
    
    mean_value = np.mean(valid_values)
    standard_deviation_value = np.std(valid_values)
    
    mean_values[column] = mean_value
    standard_deviation_values[column] = standard_deviation_value

print('Mean values for each predictor:\n', mean_values)
print('\nStandard deviation values for each predictor:\n', standard_deviation_values)
```

```python
columns_to_describe = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']

auto[columns_to_describe].describe()
```

## D

```python
subset_auto = pd.DataFrame(predictors_quantitative).T

subset_auto = subset_auto.drop(subset_auto.index[9:86])

ranges = {}
means = {}
standard_deviations = {}

for column in subset_auto.columns:

    numeric_values = pd.to_numeric(subset_auto[column], errors='coerce')
    
    valid_values = numeric_values.dropna()
    
    column_range = np.ptp(valid_values)
    mean_value = np.mean(valid_values)
    standard_deviation_value = np.std(valid_values)
    
    ranges[column] = column_range
    means[column] = mean_value
    standard_deviations[column] = standard_deviation_value

print('Range of each predictor in the subset:\n', ranges)
print('\nMean values for each predictor in the subset:\n', means)
print('\nStandard deviation values for each predictor in the subset:\n', standard_deviations)
```
